```python
import pandas as pd
import matplotlib.pyplot as plt


monthly_sales = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [12000, 15000, 17000, 16000, 18000, 21000]
})


category_sales = pd.DataFrame({
    'Category': ['Laptops', 'Mobiles', 'Accessories', 'Tablets', 'Monitors'],
    'Sales': [50000, 75000, 20000, 15000, 18000]
})


region_share = pd.Series(
    [40, 25, 20, 15],
    index=['North', 'South', 'East', 'West']
)


plt.figure(figsize=(6,4))
plt.plot(monthly_sales['Month'], monthly_sales['Sales'], marker='o', linestyle='-', color='blue')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales (in ₹)')
plt.grid(True)
plt.tight_layout()
plt.show()


plt.figure(figsize=(6,4))
plt.bar(category_sales['Category'], category_sales['Sales'], color='orange')
plt.title('Product Category Sales')
plt.xlabel('Category')
plt.ylabel('Sales (in ₹)')
plt.tight_layout()
plt.show()


plt.figure(figsize=(6,6))
colors = ['lightcoral', 'gold', 'lightgreen', 'skyblue']
region_share.plot.pie(autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Market Share by Region')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
